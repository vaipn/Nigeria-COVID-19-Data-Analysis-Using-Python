```python
fig,axes=plt.subplots(nrows=7,ncols=1,figsize=(8,38))
color=['green','orange','purple','blue']
groupd_meltd_realGDP.get_group(2014).plot(x='Year',y='GDP',ax=axes[0],kind='bar',color=color)
groupd_meltd_realGDP.get_group(2015).plot.bar(x='Year',y='GDP',ax=axes[1],color=color)
groupd_meltd_realGDP.get_group(2016).plot.bar(x='Year',y='GDP',ax=axes[2],color=color)
groupd_meltd_realGDP.get_group(2017).plot.bar(x='Year',y='GDP',ax=axes[3],color=color)
groupd_meltd_realGDP.get_group(2018).plot.bar(x='Year',y='GDP',ax=axes[4],color=color)
groupd_meltd_realGDP.get_group(2019).plot.bar(x='Year',y='GDP',ax=axes[5],color=color)
groupd_meltd_realGDP.get_group(2020).plot.bar(x='Year',y='GDP',ax=axes[6],color=color)

# Adding legend to the plot
axes[0].legend(handles=[q1,q2,q3,q4],loc='lower left')
axes[1].legend(handles=[q1,q2,q3,q4],loc='lower left')
axes[2].legend(handles=[q1,q2,q3,q4],loc='lower left')
axes[3].legend(handles=[q1,q2,q3,q4],loc='lower left')
axes[4].legend(handles=[q1,q2,q3,q4],loc='lower left')
axes[5].legend(handles=[q1,q2,q3,q4],loc='lower left')
plt.legend(handles=[q1,q2,q3],loc='lower left')

# To display numbers in y axis as is in the dataframe (and not as shortened form - scientific)
axes[0].get_yaxis().get_major_formatter().set_scientific(False)
axes[1].get_yaxis().get_major_formatter().set_scientific(False)
axes[2].get_yaxis().get_major_formatter().set_scientific(False)
axes[3].get_yaxis().get_major_formatter().set_scientific(False)
axes[4].get_yaxis().get_major_formatter().set_scientific(False)
axes[5].get_yaxis().get_major_formatter().set_scientific(False)
axes[6].get_yaxis().get_major_formatter().set_scientific(False)

# Adding an horiontal line through the graph at the value of Q2 2020.
axes[6].axhline(y=15890000.00, color='black')
```
