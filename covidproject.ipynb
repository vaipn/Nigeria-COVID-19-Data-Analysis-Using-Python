### Of the lab confirmed covid-19 cases recorded in each state, which state did the most percentage die?

```python
ten_sorted=ncdc.sort_values(by='No. of Cases (Lab Confirmed)', ascending=False)[:10]
ten_sorted['percentage']=(ten_sorted['No. of Deaths']/ten_sorted['No. of Cases (Lab Confirmed)'])*100
ten_sorted['percentage']
```

```python
norm=plt.Normalize(ten_sorted['percentage'].values.min(),ten_sorted['percentage'].values.max())
colors=plt.cm.autumn_r(norm(ten_sorted.sort_values(by='percentage')['percentage'].values))
ten_sorted.sort_values(by='percentage').plot.bar(x='States Affected',y='percentage',xlabel='States',ylabel='Percentage (%) of deaths',yticks=(np.arange(0.0, 4.5, step=0.5)),legend=False,color=colors)
```

### Edo State - 3.77%.  Of every 100 lab confirmed cases in the State, about 4 of them died.

### Which State was the most effective in providing Medical treatment to COVID-19 patients?

### TODO L - 
Determine the effect of the Pandemic on the economy. To do this, you will compare the Real GDP value Pre-COVID-19 with Real GDP in 2020 (COVID-19 Period, especially Q2 2020)
<br>
Steps
* From the Real GDP Data, generate a `barplot` using the GDP values for each year & quarters. For example: On x-axis you will have year 2017 and the bars will be values of each quarters(Q1-Q4). You expected to have subplots of each quarters on one graph.
<br>
Hint: Use [Pandas.melt](https://pandas.pydata.org/docs/reference/api/pandas.melt.html) to create your plot DataFrame 
* Set your quarter legend to lower left.
* Using `axhline`, draw a horizontal line through the graph at the value of Q2 2020.
* Write out your observation

```python
meltd_realGDP=realGDP.melt(id_vars='Year',value_name='GDP',var_name='Quarters')
```

```python
meltd_realGDP
```

```python
groupd_meltd_realGDP=meltd_realGDP.groupby('Year')
```

```python
# setting color and labels for the bars
q1=mpatches.Patch(color='green',label='Q1')
q2=mpatches.Patch(color='orange',label='Q2')
q3=mpatches.Patch(color='purple',label='Q3')
q4=mpatches.Patch(color='blue',label='Q4')
```

```python
fig,axes=plt.subplots(nrows=7,ncols=1,figsize=(8,38))
color=['green','orange','purple','blue']
groupd_meltd_realGDP.get_group(2014).plot(x='Year',y='GDP',ax=axes[0],kind='bar',color=color)
groupd_meltd_realGDP.get_group(2015).plot.bar(x='Year',y='GDP',ax=axes[1],color=color)
groupd_meltd_realGDP.get_group(2016).plot.bar(x='Year',y='GDP',ax=axes[2],color=color)
groupd_meltd_realGDP.get_group(2017).plot.bar(x='Year',y='GDP',ax=axes[3],color=color)
groupd_meltd_realGDP.get_group(2018).plot.bar(x='Year',y='GDP',ax=axes[4],color=color)
groupd_meltd_realGDP.get_group(2019).plot.bar(x='Year',y='GDP',ax=axes[5],color=color)
groupd_meltd_realGDP.get_group(2020).plot.bar(x='Year',y='GDP',ax=axes[6],color=color)

# Adding legend to the plot
axes[0].legend(handles=[q1,q2,q3,q4],loc='lower left')
axes[1].legend(handles=[q1,q2,q3,q4],loc='lower left')
axes[2].legend(handles=[q1,q2,q3,q4],loc='lower left')
axes[3].legend(handles=[q1,q2,q3,q4],loc='lower left')
axes[4].legend(handles=[q1,q2,q3,q4],loc='lower left')
axes[5].legend(handles=[q1,q2,q3,q4],loc='lower left')
plt.legend(handles=[q1,q2,q3],loc='lower left')

# To display numbers in y axis as is in the dataframe (and not as shortened form - scientific)
axes[0].get_yaxis().get_major_formatter().set_scientific(False)
axes[1].get_yaxis().get_major_formatter().set_scientific(False)
axes[2].get_yaxis().get_major_formatter().set_scientific(False)
axes[3].get_yaxis().get_major_formatter().set_scientific(False)
axes[4].get_yaxis().get_major_formatter().set_scientific(False)
axes[5].get_yaxis().get_major_formatter().set_scientific(False)
axes[6].get_yaxis().get_major_formatter().set_scientific(False)

# Adding an horiontal line through the graph at the value of Q2 2020.
axes[6].axhline(y=15890000.00, color='black')
```

#### In previous years (before COVID-19 hit the country), the GDP growth each year used to increase gradually every quarter, whereas ater the hit of COVID-19 (precisely the Second Quarter of 2020) the GDP decreased in growth.
